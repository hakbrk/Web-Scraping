```python
#Import required libraries and modules
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd
import re
import numpy as np
from tqdm import tqdm_notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
url = 'https://twitter.com/MarsWxReport?lang=en'
browser.visit(url)
browser.is_text_present('MarsWxReport', wait_time=1)
```

```python
mars_weather = []
a=0
while a < 600:
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    try:
        report_all = soup.findAll('p', class_='TweetTextSize')[a].text
    except IndexError as error:
        browser.execute_script("window.scrollTo(5, document.body.scrollHeight);")
    mars_weather.append(report_all)
    print(f'report {a} {report_all}')
    a += 1
browser.quit()
```

```python
mars_weather_data_raw = {'Raw Text' : mars_weather}
mars_data_df = pd.DataFrame(mars_weather_data_raw)
mars_data_df['Date'] = ''
mars_data_df['High Temp (F)'] = ''
mars_data_df['Low Temp (F)'] = ''
mars_data_df['Wind Direction'] = ''
mars_data_df['Wind Speed (mph)'] = ''
mars_data_df['Wind Gusting (mph)'] = ''
mars_data_df['Pressure (hPa)'] = ''
mars_data_df.head()
```

```python
mars_weather_df = mars_data_df[mars_data_df['Raw Text'].str.contains('sol' and 'low' or 'Sol' and 'low')]
```

```python
mars_weather_df.head()
```

```python
for index, row in tqdm_notebook(list(mars_weather_df.iterrows())):
    try:
        low_val = re.search('low (.*)ºF\)', mars_weather_df.loc[index, 'Raw Text']).group(1)
        mars_weather_df.loc[index, 'Low Temp (F)'] = re.search(' \((.*)ºF\)', low_val).group(1)
    except (AttributeError):
        mars_weather_df.loc[index, 'Low Temp (F)']= 'NaN'
for index, row in tqdm_notebook(list(mars_weather_df.iterrows())):
    try:
        mars_weather_df.loc[index, 'Date'] = re.search(' \((.*)\) low', mars_weather_df.loc[index, 'Raw Text']).group(1)
        high_val = re.search('ºC \((.*)ºF\)', mars_weather_df.loc[index, 'Raw Text']).group(1)
    except (AttributeError):
        mars_weather_df.loc[index, 'Date']= 'NaN'
for index, row in tqdm_notebook(list(mars_weather_df.iterrows())):
    try:
        mars_weather_df.loc[index, 'High Temp (F)'] = re.search('ºC \((.*)', high_val).group(1)
    except (AttributeError):
        mars_weather_df.loc[index, 'High Temp (F)']= 'NaN'
for index, row in tqdm_notebook(list(mars_weather_df.iterrows())):
    try:
        mars_weather_df.loc[index, 'Wind Speed (mph)'] = re.search('m/s \((.*) mph\) gusting', mars_weather_df.loc[index, 'Raw Text']).group(1)
    except (AttributeError):
        mars_weather_df.loc[index, 'Wind Speed (mph)']= 'NaN'
for index, row in tqdm_notebook(list(mars_weather_df.iterrows())):
    try:
        gust = re.search('gusting to (.*) mph\)', mars_weather_df.loc[index, 'Raw Text']).group(1)
        mars_weather_df.loc[index, 'Wind Gusting (mph)'] = re.search('m/s \((.*)', gust).group(1)
    except (AttributeError):
        mars_weather_df.loc[index, 'Wind Gusting (mph)']= 'NaN'
for index, row in tqdm_notebook(list(mars_weather_df.iterrows())):
    try:
        mars_weather_df.loc[index, 'Pressure (hPa)'] = re.search('pressure at (.*) hPa', mars_weather_df.loc[index, 'Raw Text']).group(1)
    except (AttributeError):
        mars_weather_df.loc[index, 'Pressure (hPa)']= 'NaN'
for index, row in tqdm_notebook(list(mars_weather_df.iterrows())):
    try:
        mars_weather_df.loc[index, 'Wind Direction'] = re.search('from the (.*) at', mars_weather_df.loc[index, 'Raw Text']).group(1)
    except:
        mars_weather_df.loc[index, 'Wind Direction']= 'NaN'
        
for index, row in tqdm_notebook(list(mars_weather_df.iterrows())):
    if mars_weather_df.loc[index, 'Low Temp (F)'] == 'NaN':
        try:
            low_val = re.search('low (.*), ', mars_weather_df.loc[index, 'Raw Text']).group(1)
            mars_weather_df.loc[index, 'Low Temp (F)'] = re.search('/(.*)F', low_val).group(1)
        except (AttributeError):
            mars_weather_df.loc[index, 'Low Temp (F)']= np.nan
for index, row in tqdm_notebook(list(mars_weather_df.iterrows())):
    if mars_weather_df.loc[index, 'Date'] == 'NaN':
        try:
            mars_weather_df.loc[index, 'Date'] = re.search(' \((.*)\), ', mars_weather_df.loc[index, 'Raw Text']).group(1)
        except (AttributeError):
            mars_weather_df.loc[index, 'Low Temp (F)']= np.nan
for index, row in tqdm_notebook(list(mars_weather_df.iterrows())):
    if mars_weather_df.loc[index, 'High Temp (F)'] == 'NaN':
        try:
            high_val = re.search('high (.*)F, ', mars_weather_df.loc[index, 'Raw Text']).group(1)
            mars_weather_df.loc[index, 'High Temp (F)'] = re.search('\/(.*)F,', high_val).group(1)
        except (AttributeError):
            mars_weather_df.loc[index, 'High Temp (F)']= np.nan
for index, row in tqdm_notebook(list(mars_weather_df.iterrows())):
    if mars_weather_df.loc[index, 'Wind Direction'] == 'NaN':
        try:
            mars_weather_df.loc[index, 'Wind Direction'] = re.search('from the (.*) at', mars_weather_df.loc[index, 'Raw Text']).group(1)
        except (AttributeError):
            mars_weather_df.loc[index, 'Wind Direction']= np.nan
for index, row in tqdm_notebook(list(mars_weather_df.iterrows())):
    if mars_weather_df.loc[index, 'Wind Speed (mph)'] == 'NaN':
        try:
            mars_weather_df.loc[index, 'Wind Speed (mph)'] = re.search('m/s \((.*) mph\) gusting', mars_weather_df.loc[index, 'Raw Text']).group(1)
        except (AttributeError):
            mars_weather_df.loc[index, 'Wind Speed (mph)']= np.nan            
for index, row in tqdm_notebook(list(mars_weather_df.iterrows())):
    if mars_weather_df.loc[index, 'Wind Gusting (mph)'] == 'NaN':
        try:
            gust = re.search('gusting to (.*) mph\)', mars_weather_df.loc[index, 'Raw Text']).group(1)
            mars_weather_df.loc[index, 'Wind Gusting (mph)'] = re.search('m/s \((.*)', gust).group(1)
        except (AttributeError):
            mars_weather_df.loc[index, 'Wind Gusting (mph)']= np.nan
for index, row in tqdm_notebook(list(mars_weather_df.iterrows())):
    if mars_weather_df.loc[index, 'Pressure (hPa)'] == 'NaN':
        try:
            mars_weather_df.loc[index, 'Pressure (hPa)'] = re.search('pressure at (.*) hPa', mars_weather_df.loc[index, 'Raw Text']).group(1)
        except (AttributeError):
            mars_weather_df.loc[index, 'Pressure (hPa)']= np.nan
for index, row in tqdm_notebook(list(mars_weather_df.iterrows())):
    if mars_weather_df.loc[index, 'Date'] == 'NaN':
        mars_weather_df.loc[index, 'Date'] = np.nan
```
